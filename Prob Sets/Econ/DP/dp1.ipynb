```python
##Exercise 1

import numpy as np
import matplotlib.pyplot as plt

A = [[0.6, 0.1, -0.3],
     [0.5, -0.4, 0.2],
     [1.0, -0.2, 1.1]]

b = [[12],
     [10],
     [-1]]

A, b = map(np.asarray, (A, b))


from scipy.linalg import eigvals, solve
evs = eigvals(A)
ρ = max(abs(evs))
print(ρ)

#Linear algebra solution

A1 = [[0.4, -0.1, 0.3],
     [-0.5, 1.4, -0.2],
     [-1.0, 0.2, -0.1]]
b = [[12],
     [10],
     [-1]]
A1_inv = np.linalg.inv(A1)
solution = np.dot(A1_inv, b)
print(solution)

#Successive approximations solution

x = [[0.5],
     [0.5],
     [0.5]]
i = 0
while i < 3000:
    x = np.dot(A, x) + b
    i +=1
print(x)
```
