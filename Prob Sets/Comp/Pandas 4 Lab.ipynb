```python
## Problem 6

dow = pd.read_csv('DJIA.csv')
dow.index = pd.to_datetime(dow['DATE'])
dow['VALUE'] = dow['VALUE'].replace({'.': np.nan})
dow[['VALUE']] = dow[['VALUE']].apply(pd.to_numeric)
del dow['DATE']

dow['PREV'] = dow['VALUE']
dow.PREV = dow.PREV.shift(1)
dow['DIFF'] = dow['VALUE'] - dow['PREV']

print(dow)

#Single day with the largest gains
print(dow.loc[dow['DIFF'].idxmax()])
print("")

#Single day with the largest loss
print(dow.loc[dow['DIFF'].idxmin()])
print("")

#Month with the largest gain
dow.sort_index(inplace=True)
dow = dow.resample('M').first()
dow['PREV'] = dow['VALUE']
dow.PREV = dow.PREV.shift(1)
dow['DIFF'] = dow['VALUE'] - dow['PREV']

print(dow.loc[dow['DIFF'].idxmax()])
print("")

#Month with the largest loss
print(dow.loc[dow['DIFF'].idxmin()])
print("")

```

The single day with the largest gains was October 13, 2008. The single day with the largest losses was September 29, 2008. The single month with the largest gains was November 2015. The single month with the largest losses was September 2015.

```python
## Problem 7

from matplotlib import pyplot as plt

dow = pd.read_csv('DJIA.csv')
dow.index = pd.to_datetime(dow['DATE'])
dow['VALUE'] = dow['VALUE'].replace({'.': np.nan})
dow[['VALUE']] = dow[['VALUE']].apply(pd.to_numeric)
del dow['DATE']
dow=dow.dropna(axis=0, how='any')


dow['ROL30'] = dow.rolling(window=30).mean().values
dow['ROL365'] = dow['VALUE'].rolling(window=365).mean().values
dow['EXP30'] = pd.ewma(dow["VALUE"], span=30, freq="D")
dow['EXP30'] = pd.ewma(dow["VALUE"], span=365, freq="D")
print(dow)


plt.plot(dow)
plt.legend()

plt.title("Dow Jones Industrial Average")
plt.xlabel("Date")
plt.ylabel("DJIA")
plt.show()
```
