```python
## Problem 1

import random
import numpy as np
from matplotlib import pyplot as plt

def normal_matrix(n):
    return np.random.randn(n,n)

def row_means(matrix):
    return np.mean(matrix, axis=1)

def numpy_var(vals):
    return np.var(vals)

n = 100
n_vals = []
variances = []
while n <=1000:
    matrix = normal_matrix(n)
    means = row_means(matrix)
    variance = numpy_var(means)
    variances.append(variance)
    n_vals.append(n)
    n+=100
    
plot1 = plt.plot(n_vals, variances)
plt.show()
    
```

```python
## Problem 4

x = np.linspace(-2*np.pi, 2*np.pi, 100)

plt.subplot(221)
plt.plot(x, np.sin(x), 'g')
plt.axis([0,2*np.pi,-2,2])
plt.title("f(x) = sin(x)")

plt.subplot(222)
plt.plot(x, np.sin(2*x),'r--')
plt.axis([0,2*np.pi,-2,2])
plt.title("f(x) = sin(2x)")

plt.subplot(223)
plt.plot(x, 2*np.sin(x),'b--')
plt.axis([0,2*np.pi,-2,2])
plt.title("f(x) = 2sin(x)")

plt.subplot(224)
plt.plot(x, 2*np.sin(2*x),'m:')
plt.axis([0,2*np.pi,-2,2])
plt.title("f(x) = 2sin(2x)")

plt.suptitle("Fun with Sine")
plt.tight_layout()

plt.show()
```

```python
## Problem 5

data = np.load('FARS.npy')
rows = np.transpose(data)


plt.subplot(121)
plt.plot(rows[1],rows[2],'k,')
plt.axis('equal')
plt.title("Locations of Fatal Injuries")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.subplot(122)
plt.hist(rows[0], bins=24)
plt.xlim(0,24)
plt.title("Fatal Injuries by Hour of Day")
plt.xlabel("Hour")

plt.tight_layout()
plt.show()

```

```python
## Problem 6

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.copy(x)
X, Y = np.meshgrid(x,y)
Z = (np.sin(X)*np.sin(Y))/(X*Y)

plt.subplot(121)
plt.pcolormesh(X, Y, Z, cmap = "viridis")
plt.colorbar()
plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-2*np.pi, 2*np.pi)
plt.show()

plt.subplot(122)
plt.contourf(X, Y, Z, cmap = "viridis")
plt.colorbar()
plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-2*np.pi, 2*np.pi)
plt.show()
```
