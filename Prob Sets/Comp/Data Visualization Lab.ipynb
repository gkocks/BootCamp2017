```python
## Problem 1

import numpy as np
from matplotlib import pyplot as plt

anscombe = np.load('anscombe.npy')
rows = np.transpose(anscombe)

x = np.linspace(0,20,10)
y = 0.5*x + 3

plt.subplot(221)
plt.plot(rows[0], rows[1], 'o')
plt.axis([0,20,0,13])
plt.plot(x,y)

plt.subplot(222)
plt.plot(rows[2], rows[3], 'o')
plt.axis([0,20,0,13])
plt.plot(x,y)

plt.subplot(223)
plt.plot(rows[4], rows[5], 'o')
plt.axis([0,20,0,13])
plt.plot(x,y)

plt.subplot(224)
plt.plot(rows[6], rows[7], 'o')
plt.axis([0,20,0,13])
plt.plot(x,y)

plt.suptitle("Anscombe's quartet")
plt.show()
```

The first set of data is approximately linear but scattered a little bit around the line of best fit. The second set of data points neatly follows a quadratic function. The third set of data points is exactly linear with a slope less than 1/2, but with one strong outlier. The final set of data points is al clustered at x=8 with a small range of y values, and then one strong outlier with a much larger x. Overall, each set of data points has similar summary statistics but the visualizations show signficant differences in the features of the data.

```python
## Problem 2
import scipy.misc

x = np.linspace(-1, 1, 200)

n = 0
while n<=3:
    v=0
    while v<=n:
        plt.subplot(4,4,4*n+v+1)
        plt.plot(x, scipy.misc.comb(n,v)*(x**v)*(1-x)**(n-v))
        plt.title("v=" + str(v) + ", n=" + str(n))
        plt.axis([-1.0, 1.0, 0.0, 4.0])
        plt.xticks([-1.0, 0.0, 1.0])
        plt.yticks([0.0, 2.0, 4.0])
        #plt.axis([0,1,0,1])
        if 4*n+v+1 < 13:
            plt.tick_params(labelbottom="off")
        if v != 0:
            plt.tick_params(labelleft = "off")
        v += 1
    n+=1

    
plt.suptitle("Bernstein Basis Polynomials")
plt.tight_layout()
plt.show()
```

```python
## Problem 3

mlb = np.load('MLB.npy')
rows = np.transpose(mlb)
height = rows[0]
weight = rows[1]
age = rows[2]

plt.scatter(height, weight, c=age)
cbar = plt.colorbar()
plt.xlabel("Height (inches)")
plt.ylabel("Weight (pounds)")
cbar.set_label("Age (years)")
plt.grid()
plt.title("Height and Weight of MLB Players")

#Determine the regression equation
m,b = np.polyfit(height, weight, 1)
x = np.linspace(66,84,100)
y = m*x+b
plt.plot(x,y)

plt.show()
```

```python
## Problem 4

earthquakes = np.load('earthquakes.npy')
rows = np.transpose(earthquakes)
year = rows[0]
magnitude = rows[1]
longitude = rows[2]
latitude = rows[3]

plt.subplot(221)
plt.hist(year, bins=[2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010])
plt.xlabel("year")
plt.ylabel("number")
plt.title("Earthquakes per year")

plt.subplot(222)
plt.hist(magnitude, bins=30)
plt.xticks([5.0,6.0,7.0,8.0,9.0])
plt.xlabel("magnitude")
plt.ylabel("number")
plt.title("Earthquakes by strength")

plt.subplot(223)
plt.scatter(longitude, latitude, c=magnitude, s=10, alpha=.5)
cbar = plt.colorbar()
plt.xlabel("Longitude")
plt.xticks([-150, -50, 50, 150])
plt.yticks([-60, -20, 20, 60, 100])
plt.ylabel("Latitude")
cbar.set_label("Magnitude")
plt.grid()
plt.title("Strength of Earthquakes by Location")



plt.tight_layout()
plt.show()
```

```python
## Problem 5

from matplotlib.colors import LogNorm
x = np.linspace(-2, 4, 200)
X, Y = np.meshgrid(x,x.copy())
Z = (1-X)**2 + 100*(Y-X**2)**2

plt.pcolormesh(X, Y, Z, cmap="plasma", norm=LogNorm())
plt.colorbar()

x = np.array([1])
y = np.array([1])
plt.plot(x,y,"or")

plt.title("Rosenbrock Function")

plt.show()
```
